```python
#Scatter plot - To print Scatter plot charts

#df.plot.scatter(x='Texture Rating',y='Overall Rating') #need to mention the 'x' and 'y' axis
df.plot.scatter(x='Texture Rating',y='Overall Rating',s = 500, c = 'red') #'s'-size and 'c'- colour
```

```python
#Histogram - to make histogram chart
#good for showing distributions of the variables.

df.plot.hist() #normal
df.plot.hist(bins = 20) #default is 10 , mentions how may tickmarks on x - axis.

```

```python
#Box plot - to make Box plot chart
#it tell lower minimum , 50 % , higher minimum.

df.boxplot() #normal
```

```python
#Area plot - To print Area plot chart

#df.plot.area() #normal

df.plot.area(figsize = (10,5)) #to change the size of the fig.
```

```python
#Pie chart - to print pie chart
#have to specify column name working with.

#df.plot.pie(y ='Texture Rating') #mentioned column name
df.plot.pie(y ='Texture Rating',figsize = (10,6)) to set the size
```
