```python
#Merge,Join & Concatenate

#how='inner' - natural/inner join 
#how='outer' - full outer join 
#how='left' - left outer join 
#how='right' = right outer join 

import pandas as pd
```

```python
df1 = pd.read_csv(r"C:\Users\Ganseh\Downloads\LOTR.CSV")
df1 #first data frame
```

```python
df2 = pd.read_csv(r"C:\Users\Ganseh\Downloads\LOTR 2.CSV")
df2 #second data frame
```

```python
#innner join is default for merge

#df1.merge(df2) #by default it is 'inner join'

#df1.merge(df2, how = 'inner', on = ['FellowshipID']) #it will make two column for two data sets

df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName']) #ti will give inner join (common elements)
```

```python
#outer join - all the values

df1.merge(df2, how="outer") #returns all values and 'Nan' for null values.
```

```python
#left outer join - all result in left data set and common between left and right

df1.merge(df2, how="left")
```

```python
#right outer join - all results from right data set and common between right and left.


df1.merge(df2, how="right")
```

```python
#interview and not for regular work
#it compares the values in left data frame to the values in the right data frame.
df1.merge(df2, how="cross")
```

```python
#JOIN() it is used to join the two data sets together.
#we have to mention the suffix.

#here we have to modify the 'left data set suffix' and 'right data set' suffix
#so that it can differentiate between them. 

#it works better with the indexes.


df1.join(df2, on = 'FellowshipID', how = 'outer' , lsuffix='_left',rsuffix='_right')
```

```python
# setting indexex using the joins . here we can set up the columns to be taken 
#as the indexes for the result. and also can mention the type of the joins.

df3 = df1.set_index("FellowshipID").join(df2.set_index("FellowshipID"),lsuffix='_left', rsuffix='_right',how = 'outer')
df3

#here we have set the name of the left and right data set as '_left' and '_right' by using lsuffix and rsuffix 
#and we have also mentioned the join type by using the 'how' function.

#'merge' can do it in less time bu 'join' is better at indexing
```

```python
#concat - puts one date frame on the other data frame
#it does not join by default but even after giving join type it takes only common columns
#axis - by mentioning this function we can join data sets on the index.

pd.concat([df1,df2],join='outer',axis = 1)

#used axis for index setting.
```

```python
#not recommended #outdated un operational

#Append() - it used to append the rows of one data frame to the end of another data frame
# and then  we can return it as another data frame.

df1.append(df2) 
```
